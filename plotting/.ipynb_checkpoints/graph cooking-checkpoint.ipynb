```python
import matplotlib.pyplot as plt
```

```python
byron = ['ByronTalbott', 158.9, 129.9]
donal = ['Donal Skehan', 1072, 78.76]
hart = ['MyHarto',22257,4053]
laura = ['LauraVitalesKitchen',2875.79,163.26]

babish = ['Binging with Babish',7814.8,1664.84]
dog = ['Cooking with Dog', 5638.9,1398.6]
peaceful = ['Peaceful Cuisine',3346.27,302.15]
english = ['English Heritage',4555.5,622.18]

manjula = ['Manjulas Kitchen',278.77,2.8]
burger = ['The Burger Show',4143.21,1098.85]
chef = ['ChefSteps',624.4,148.0]
yousuck = ['You Suck At Cooking',5000.4,1276.4]
```

```python
fig,ax = plt.subplots(2,2,figsize = (6,7))
plt.subplots_adjust(right=1.7)
names = [byron[0],donal[0],hart[0],laura[0]]
positive_score = [byron[1],donal[1],hart[1],laura[1]]
negative_score = [byron[2],donal[2],hart[2],laura[2]]

for i in range(4):
    if i<2:
        ax[0][i].pie([positive_score[i],negative_score[i]],autopct='%1.1f%%',shadow=False)
        ax[0][i].axis('equal')
        ax[0][i].set_title(names[i])
        ax[0][i].legend(('pos','neg'))
    else:
        ax[1][i-2].pie([positive_score[i],negative_score[i]],autopct='%1.1f%%',shadow=False)
        ax[1][i-2].axis('equal')
        ax[1][i-2].set_title(names[i])
        ax[1][i-2].legend(('pos','neg'))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

fileName = '../sentiment_analysis/textblob_data/cooking.txt'    
```

```python
#read in data
df = pd.read_csv(fileName, sep=" ",header=None)
df.columns = ['Creator','Positive','Negative']
#fix Creator Names
df['Creator'] = ['ByronTalbott','Donal Skehan','MyHarto','LauraVitalesKitchen','Binging with Babish','Cooking with Dog','Peaceful Cuisine','English Heritage','Manjulas Kitchen','The Burger Show','ChefSteps','You Suck At Cooking']
df.set_index('Creator')

#Normalize Scores
df['Negative'] = abs(df['Negative'])
df['Normalized Positive'] = df['Positive']/(df['Positive']+df['Negative'])
df['Normalized Negative'] = df['Negative']/(df['Positive']+df['Negative'])
```

```python
#bar plot
df.plot.bar(x='Creator',y=['Normalized Positive','Normalized Negative'])
```
