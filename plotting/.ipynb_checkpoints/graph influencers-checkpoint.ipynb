```python
import pandas as pd
import matplotlib.pyplot as plt

fileName = '../sentiment_analysis/textblob_data/influencers.txt'    
```

```python
### read in data
df = pd.read_csv(fileName, sep=" ",header=None)
df.columns = ['Creator','Positive','Negative']
df.set_index('Creator', inplace=True)

#Normalize Scores
df['Negative'] = abs(df['Negative'])
df['Normalized Positive'] = df['Positive']/(df['Positive']+df['Negative'])*100
df['Normalized Negative'] = df['Negative']/(df['Positive']+df['Negative'])*100
```

```python
df['Gender'] = ['Male', 'Male', 'Male', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male','Male', 'Female', 'Female']
```

```python
df['Category'] = ['Influencers']*len(df)
```

```python
df
```

```python
fileName2 = '../sentiment_analysis/textblob_data/cooking.txt'    
```

```python
#read in data
df2 = pd.read_csv(fileName2, sep=", ",header=None)
df2.columns = ['Creator','File','Positive','Negative','Gender','Country']
df2.set_index('Creator', inplace=True)

#Normalize Scores
df2['Negative'] = abs(df2['Negative'])
df2['Normalized Positive'] = df2['Positive']/(df2['Positive']+df2['Negative'])*100
df2['Normalized Negative'] = df2['Negative']/(df2['Positive']+df2['Negative'])*100
```

```python
df2.drop(columns=['File', 'Country'], inplace=True)
```

```python
df2['Category'] = ['Cooking']*len(df2)
```

```python
df2
```

```python
### read in data
fileName3 = '../sentiment_analysis/textblob_data/gaming.txt'    
df3 = pd.read_csv(fileName3, sep=" ",header=None)
df3.columns = ['Creator','Positive','Negative']
df3.set_index('Creator', inplace=True)

#Normalize Scores
df3['Negative'] = abs(df3['Negative'])
df3['Normalized Positive'] = df3['Positive']/(df3['Positive']+df3['Negative'])*100
df3['Normalized Negative'] = df3['Negative']/(df3['Positive']+df3['Negative'])*100
```

```python
df3['Category'] = ['Gaming']*len(df3)
```

```python
df3['Gender'] = ['Male']*len(df3)
```

```python
frames = [df, df2, df3]
```

```python
result = pd.concat(frames)
```

```python
g = result.groupby(by='Gender')
```

```python
male = g.get_group('Male')
```

```python
female = g.get_group('Female')
```

```python
#vs category
d2 = result.groupby(by=['Gender','Category']).mean().B.count().unstack().reset_index().plot.bar(x='Category', y=['Male', 'Female'])

ax = d2.plot.bar(y=['Normalized Negative'],width = 0.8,rot=45)
ax.set_xticklabels(ax.get_xticklabels(),ha='right');
ax.legend(loc='upper left', bbox_to_anchor=(1,1));
```

```python
d2
```

```python
d2 = result.groupby(by=['Gender','Category']).mean()['Normalized Negative'].unstack().reset_index()
d2.set_index('Gender', inplace=True)
```

```python
ax = d2.plot.bar(y=['Gaming', 'Cooking', 'Influencers'],width = 0.8,rot=45)
ax.set_xticklabels(ax.get_xticklabels(),ha='right')
ax.legend(loc='upper left', bbox_to_anchor=(1,1))
ax.set_ylabel('% of Negative Comments')
```

```python
#positive-negative bar plot
ax = df.plot.bar(x='Creator',y=['Normalized Positive','Normalized Negative'],width=0.8,rot=45)
ax.set(xlabel = '', ylabel = 'Percent');
ax.set_xticklabels(ax.get_xticklabels(),ha='right');
ax.legend(loc='best');
```

```python
#vs gender
d2 = df.groupby(by='Gender').mean()

ax = d2.plot.bar(y=['Normalized Positive','Normalized Negative'],width = 0.8,rot=45)
ax.set_xticklabels(ax.get_xticklabels(),ha='right');
ax.legend(loc='upper left', bbox_to_anchor=(1,1));
```
