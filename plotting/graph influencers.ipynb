```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
### read in data
fileName = '../sentiment_analysis/textblob_data/influencers.txt'    
influencers = pd.read_csv(fileName, sep=" ",header=None)
influencers.columns = ['Creator','Positive','Negative']
c = list(influencers['Creator'])
c = [i.split('.')[0] for i in c]
influencers['Creator'] = c

influencers.set_index('Creator', inplace=True)

#Normalize Scores
influencers['Negative'] = abs(influencers['Negative'])
influencers['Normalized Positive'] = influencers['Positive']/(influencers['Positive']+influencers['Negative'])*100
influencers['Normalized Negative'] = influencers['Negative']/(influencers['Positive']+influencers['Negative'])*100
```

```python
influencers['Gender'] = ['Male', 'Male', 'Male', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male','Male', 'Female', 'Female']
```

```python
influencers['Category'] = ['Influencers']*len(influencers)
```

```python
influencers
```

```python
#read in data
fileName2 = '../sentiment_analysis/textblob_data/cooking.txt'    
cooking = pd.read_csv(fileName2, sep=", ",header=None)
cooking.columns = ['Creator','File','Positive','Negative','Gender','Country']
c = list(cooking['File'])
c = [i.split('.')[0] for i in c]
cooking['Creator'] = c
cooking.set_index('Creator', inplace=True)

#Normalize Scores
cooking['Negative'] = abs(cooking['Negative'])
cooking['Normalized Positive'] = cooking['Positive']/(cooking['Positive']+cooking['Negative'])*100
cooking['Normalized Negative'] = cooking['Negative']/(cooking['Positive']+cooking['Negative'])*100
```

```python
cooking.drop(columns=['File', 'Country'], inplace=True)
```

```python
cooking['Category'] = ['Cooking']*len(cooking)
```

```python
cooking
```

```python
### read in data
fileName3 = '../sentiment_analysis/textblob_data/gaming.txt'    
gaming = pd.read_csv(fileName3, sep=" ",header=None)
gaming.columns = ['Creator','Positive','Negative']
c = list(gaming['Creator'])
c = [i.split('.')[0] for i in c]
gaming['Creator'] = c
gaming.set_index('Creator', inplace=True)

#Normalize Scores
gaming['Negative'] = abs(gaming['Negative'])
gaming['Normalized Positive'] = gaming['Positive']/(gaming['Positive']+gaming['Negative'])*100
gaming['Normalized Negative'] = gaming['Negative']/(gaming['Positive']+gaming['Negative'])*100
```

```python
gaming['Category'] = ['Gaming']*len(gaming)
```

```python
gaming['Gender'] = ['Male']*len(gaming)
```

```python
gaming
```

```python
frames = [cooking, gaming, influencers]
```

```python
result = pd.concat(frames)
```

```python
result
```

```python
g = result.groupby(by='Gender')
```

```python
male = g.get_group('Male')
```

```python
female = g.get_group('Female')
```

```python
#vs category
d2 = result.groupby(by=['Category']).mean()

ax = d2.plot.bar(y=['Normalized Negative'],width = 0.8,rot=45)
ax.set_xticklabels(ax.get_xticklabels(),ha='right');
ax.set_ylabel('% of Negative Comments')
ax.legend(loc='upper left', bbox_to_anchor=(1,1));
ax.get_legend().remove()
```

```python
#vs category
d2 = result.groupby(by=['Gender']).mean()

ax = d2.plot.bar(y=['Normalized Negative'],width = 0.8,rot=45)
ax.set_xticklabels(ax.get_xticklabels(),ha='right');
ax.set_ylabel('% of Negative Comments')
ax.legend(loc='upper left', bbox_to_anchor=(1,1));
ax.get_legend().remove()
```

```python
d2
```

```python
d2 = result.groupby(by=['Gender','Category']).mean()['Normalized Negative'].unstack().reset_index()
d2.set_index('Gender', inplace=True)
```

```python
ax = d2.plot.bar(y=['Gaming', 'Cooking', 'Influencers'],width = 0.8,rot=45)
ax.set_xticklabels(ax.get_xticklabels(),ha='right')
ax.legend(loc='best')
ax.set_ylabel('% of Negative Comments')
```
