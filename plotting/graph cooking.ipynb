```python
import pandas as pd
import matplotlib.pyplot as plt

fileName = '../sentiment_analysis/textblob_data/cooking.txt'    
```

```python
#read in data
df = pd.read_csv(fileName, sep=",",header=None)
df.columns = ['Creator','File','Positive','Negative','Gender','Country']
df.set_index('Creator')

#Normalize Scores
df['Negative'] = abs(df['Negative'])
df['Normalized Positive'] = df['Positive']/(df['Positive']+df['Negative'])*100
df['Normalized Negative'] = df['Negative']/(df['Positive']+df['Negative'])*100
```

```python
#positive-negative bar plot
ax = df.plot.bar(x='Creator',y=['Normalized Positive','Normalized Negative'],width=0.8,rot=45)
ax.set(xlabel = '', ylabel = 'Percent');
ax.set_xticklabels(ax.get_xticklabels(),ha='right');
ax.legend(loc='best');
```

```python
#vs gender
d2 = df.groupby(by='Gender').sum()
d2['Normalized Positive'] = d2['Positive']/(d2['Positive']+d2['Negative'])*100
d2['Normalized Negative'] = d2['Negative']/(d2['Positive']+d2['Negative'])*100

ax = d2.plot.bar(y=['Normalized Positive','Normalized Negative'],width = 0.8,rot=45)
ax.set_xticklabels(ax.get_xticklabels(),ha='right');
ax.legend(loc='best');
```

```python
#vs country
d2 = df.groupby(by='Country').sum()
d2['Normalized Positive'] = d2['Positive']/(d2['Positive']+d2['Negative'])*100
d2['Normalized Negative'] = d2['Negative']/(d2['Positive']+d2['Negative'])*100

ax = d2.plot.bar(y=['Normalized Positive','Normalized Negative'],width = 0.8,rot=45)
ax.set_xticklabels(ax.get_xticklabels(),ha='right');
ax.legend(loc='best');
```
