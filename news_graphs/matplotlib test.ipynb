```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig,ax = plt.subplots()
names = ['news1','news2','news3']
positive_score = [50,30,20]
negative_score = [10,80,40]
ax.barh(names,positive_score,height=0.2,align='edge')
ax.barh(names,negative_score,height=-0.2,align='edge')
ax.set_title('How each news channel reacted to Birth of Royal Baby')
ax.legend(('pos','neg'),loc='best')
#ax.spines['right'].set_color('none')
#ax.spines['top'].set_color('none')
ax.set_xlabel('sentiment value')
```

```python
fig,ax = plt.subplots(1,2,figsize = (7,15))
fig.suptitle('this is a giant title',va='center',ha='center')
plt.subplots_adjust(right=1.7)
names = ['news1','news2','news3']
positive_score = [50,30,20]
negative_score = [10,80,40]
#plot in one subplot
ax[0].pie(negative_score,labels=names,autopct='%1.1f%%',shadow=False,startangle=80) #startangle rotates the graph
ax[0].axis('equal') #makes sure that the pie chart is circular
ax[0].set_title('How negative each channel was to XXX')

#plot in other subplot
ax[1].pie(positive_score,labels=names,autopct='%1.1f%%',shadow=False,startangle=80)
ax[1].axis('equal')
ax[1].set_title('How positive each channel was to XXX')
```
