```python
def pos_neg_sent(file):
    assert isinstance(file,str)
    pos = []
    neg = []
    normalized_pos = [] #normalized positive score
    normalized_neg = [] #normalized negative score
    with open(file,'r') as f:
        a = f.readlines()
        for i in range(len(a)):
            data = a[i].split(' ')
            if len(data) == 4:
                #get the positive and negative polarities for each video category
                val1 = float(data[2])
                val2 = abs(float(data[3]))
                pos.append(val1)
                neg.append(val2)
                total = val1 + val2
                if total == 0:
                    total = 1
                normalized_pos.append(val1/total)
                normalized_neg.append(val2/total)
            else:
                val1 = float(data[1])
                val2 = abs(float(data[2]))
                pos.append(val1)
                neg.append(val2)
                total = val1 + val2
                if total == 0:
                    total = 1
                normalized_pos.append(val1/total)
                normalized_neg.append(val2/total)
            
    return pos,neg,normalized_pos,normalized_neg
```

```python
def sum_polarities(file):
    assert isinstance(file,str)
    pos_sum = 0
    neg_sum = 0
    with open(file,'r') as f:
        a = f.readlines()
        for i in range(len(a)):
            data = a[i].split(' ')
            pos_sum += float(data[-2])
            neg_sum += abs(float(data[-1]))
    return pos_sum,neg_sum
```

```python
abc_pos_sum,abc_neg_sum = sum_polarities('ABC2.txt')
bbc_pos_sum,bbc_neg_sum = sum_polarities('BBC2.txt')
cbs_pos_sum,cbs_neg_sum = sum_polarities('CBS2.txt')
cnn_pos_sum,cnn_neg_sum = sum_polarities('CNN2.txt')
fox_pos_sum,fox_neg_sum = sum_polarities('FOX2.txt')
msnbc_pos_sum,msnbc_neg_sum = sum_polarities('MSNBC2.txt')
```
