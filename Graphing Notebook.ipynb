```python
import pandas as pd
from plotting import category_graph_functions as cgf
from plotting import news_graph_functions as ngf
import os
```

```python
### Data Initialization
data_path = os.path.join('.','sentiment_analysis','textblob_data')
sub_path = os.path.join('.','scraping')
cooking = cgf.get_cooking_data(data_path)
gaming = cgf.get_gaming_data(data_path)
influencers = cgf.get_influencer_data(data_path)
cgf.normalize_sentiment(cooking)
cgf.normalize_sentiment(gaming)
cgf.normalize_sentiment(influencers)
result = pd.concat([cooking, gaming, influencers], sort=False)
with_subs = cgf.get_subscriber_counts(sub_path, result)
```
